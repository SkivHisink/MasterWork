```python
# Ð¡Ð¾Ð·Ð´Ð°Ð½Ð¸Ðµ Ð¸ Ð¾Ð±ÑƒÑ‡ÐµÐ½Ð¸Ðµ Ð¼Ð¾Ð´ÐµÐ»Ð¸ ARCH
model = arch_model(returns)
results = model.fit()
```

```python
# ÐŸÑ€Ð¾Ð³Ð½Ð¾Ð· Ð½Ð° 30 Ð´Ð½ÐµÐ¹ Ð²Ð¿ÐµÑ€ÐµÐ´
forecast_horizon = 30
forecasts = results.forecast(horizon=forecast_horizon)
```
