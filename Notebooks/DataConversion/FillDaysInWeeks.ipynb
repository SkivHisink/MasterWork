```python
import pandas as pd
import matplotlib as plt
import numpy as np
```

```python
#load data
data = pd.read_csv("..\..\Data\Day\S&P 500 Historical Data00-20.csv")
```

```python
#transform data
data = data.replace(",", "", regex=True)
data.Date = pd.to_datetime(data.Date)
data = data.sort_values(by=["Date"])
data.set_index('Date')
data.Price = data.Price.astype(float)
data.High = data.High.astype(float)
data.Low = data.Low.astype(float)
data.Open = data.Open.astype(float)
data["Vol."] = data["Vol."].astype(float)
```

```python
data.head()
```

```python
data.index = data.index[::-1]
```

```python
data.head()
```

```python
data.set_index('Date', inplace=True)
```

```python
data.head(10)
```

```python
data.interpolate(method='time', inplace=True)
```

```python
data.head(10)
```

```python
# Создайте новый DataFrame с полным набором дат
all_dates = pd.date_range(start=data.index.min(), end=data.index.max(), freq='D')
new_df = pd.DataFrame(index=all_dates)

# Объедините новые даты со значениями из исходного DataFrame
new_df = new_df.merge(data, how='left', left_index=True, right_index=True)

# Примените линейную интерполяцию к пропущенным значениям
#new_df.interpolate(method='linear', inplace=True)
# Заполните пропущенные значения константной интерполяцией
new_df.fillna(method='ffill', inplace=True)  # Используйте 'ffill' для прямой интерполяции
```
