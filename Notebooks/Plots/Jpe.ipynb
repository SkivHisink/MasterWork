```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Загрузка данных из CSV-файла
day_data = pd.read_csv("..\..\Data\Day\jpe_Linear.csv")
week_data = pd.read_csv("..\..\Data\Week\jpe_Linear.csv")
month_data = pd.read_csv("..\..\Data\Month\jpe_Linear.csv")
```

```python
# Построение диаграммы рассеяния
plt.figure(figsize=(8, 6))
plt.plot(day_data['Date'], day_data['Close'], label='Ежедневные данные')
plt.plot(week_data['Date'], week_data['Close'], label='Еженедельные данные')
plt.plot(month_data['Date'], month_data['Close'], label='Ежемесячные данные')
plt.xlabel('Дата')
plt.ylabel('Цена')

# Уменьшение количества меток на оси x
plt.xticks(rotation=45, ha='right')
plt.xticks(day_data['Date'][::len(day_data) // 10])

plt.legend()
plt.show()
```
